```python
SIZE = 22
SMALL_SIZE=18
plt.rc('font', size=SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SIZE)    # legend fontsize
plt.rc('figure', titlesize=SIZE)  # fontsize of the figure title

with plt.xkcd():
    # Based on "Stove Ownership" from XKCD by Randall Munroe
    # https://xkcd.com/418/
    
    fig, axs = plt.subplots(1, 3, figsize=(20,8), dpi=100)
    
    
    for ax in axs:
        #ax.set(xlabel='x-label', ylabel='y-label')
        #ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
        
        ax.spines.right.set_color('none')
        ax.spines.top.set_color('none')
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_ylim([0.2, 0.7])
        ax.set_xlabel('publication year')
        ax.set_ylabel('Open Access rate (%)')


    data_flat = [0.4, 0.39, 0.41, 0.41, 0.4]
    data_increase = [0.3, 0.35, 0.4, 0.45, 0.55]
    data_inverted = [0.3, 0.35, 0.4, 0.45, 0.35]


    #ax.plot(data_flat)
    #ax.plot(data_increase)
    axs[0].plot(data_increase)
    axs[0].set_title('Increasing curve')
    
    for i in [1,2,3]:
        axs[0].annotate(
            'Open Access rates\nall measured at the\nsame time (with a\nsnapshot of the\nUnpaywall database)',
            xy=(i, data_increase[i]), arrowprops=dict(arrowstyle='->'), xytext=(1, 0.5), fontsize=SMALL_SIZE)
    

    
    axs[1].plot(data_flat)
    axs[1].set_title('Flat curve')
    
    axs[2].plot(data_inverted)
    axs[2].set_title('Inverted-V curve')
    
    axs[2].annotate(
            'The most recent publications\nare not the most open\nin the inverted-V curve case',
            xy=(4, data_inverted[4]), arrowprops=dict(arrowstyle='->'), xytext=(1, 0.6), fontsize=SMALL_SIZE)
    
    fig.savefig('types_curve.png')

    
   
```
