```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
SIZE = 22
SMALL_SIZE=18
plt.rc('font', size=SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SIZE)    # legend fontsize
plt.rc('figure', titlesize=SIZE)  # fontsize of the figure title

with plt.xkcd():
    # Based on "Stove Ownership" from XKCD by Randall Munroe
    # https://xkcd.com/418/
    
    fig, axs = plt.subplots(1, 3, figsize=(20,8), dpi=100)
    
    
    for ax in axs:
        #ax.set(xlabel='x-label', ylabel='y-label')
        #ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
        
        ax.spines.right.set_color('none')
        ax.spines.top.set_color('none')
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_ylim([0.2, 0.7])
        ax.set_xlabel('publication year')
        ax.set_ylabel('Open Access rate (%)')


    data_flat = [0.4, 0.39, 0.41, 0.41, 0.4]
    data_increase = [0.3, 0.35, 0.4, 0.45, 0.55]
    data_inverted = [0.3, 0.35, 0.4, 0.45, 0.35]


    #ax.plot(data_flat)
    #ax.plot(data_increase)
    axs[0].plot(data_increase)
    axs[0].set_title('Increasing curve')
    
    for i in [1,2,3]:
        axs[0].annotate(
            'Open Access rate\nall measured at the\nsame time (with a\nsnapshot of the\nUnpaywall database)',
            xy=(i, data_increase[i]), arrowprops=dict(arrowstyle='->'), xytext=(1, 0.5), fontsize=SMALL_SIZE)
    

    
    axs[1].plot(data_flat)
    axs[1].set_title('Flat curve')
    
    axs[2].plot(data_inverted)
    axs[2].set_title('Inverted curve')
    
    axs[2].annotate(
            'The most recent publications\nare not the most open\nin the inverted curbe case',
            xy=(4, data_inverted[4]), arrowprops=dict(arrowstyle='->'), xytext=(1, 0.6), fontsize=SMALL_SIZE)
    
    fig.savefig('types_curve.png')

    
   
```

```python
with plt.xkcd():
    # Based on "Stove Ownership" from XKCD by Randall Munroe
    # https://xkcd.com/418/

    fig = plt.figure(figsize=(15,8), dpi=100)

    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.spines.right.set_color('none')
    ax.spines.top.set_color('none')

    ax.set_yticks([])
    ax.set_ylim([0.2, 0.7])

    data1 = [0.3, 0.35, 0.40, 0.32]
    data2 = [0.43, 0.43, 0.45, 0.46, 0.43]
    data3 = [0.45, 0.48, 0.5, 0.53, 0.55, 0.6]


    # Set number of ticks for x-axis
    ax.set_xticks(range(0, len(data3)))
    # Set ticks labels for x-axis
    ax.set_xticklabels(['Y-6', 'Y-5', 'Y-4', 'Y-3', 'Y-2', 'Y-1']
                        , rotation='vertical', fontsize=SMALL_SIZE)

    ax.plot(data1, label='Observed in Y-2')
    ax.plot(data2, label='Observed in Y-1')
    ax.plot(data3, label='Observed in Y')
    
    ax.legend()

    ax.annotate(
            'Last publication\nyear Y-1 is observed in Y',
            xy=(5, data3[5]), arrowprops=dict(arrowstyle='->'), xytext=(4, 0.47), fontsize=SMALL_SIZE)

    ax.set_xlabel('publication year')
    ax.set_ylabel('Open Access rate (%)')
    
    ax.set_title('Open access curve shape dynamics')
    
    fig.savefig('curve_dynamic.png')
```

```python
with plt.xkcd():

    
    fig, axs = plt.subplots(1, 2, figsize=(20,8), dpi=100)
    
    
    width = 0.25

    oa1 = [0.15, 0.16, 0.17, 0.17]
    oa2 = [0.13, 0.14, 0.17, 0.12]
    oa3 = [0.2, 0.2, 0.25, 0.25]

    labels = ['Y-4', 'Y-3', 'Y-2', 'Y-1']
    
    axs[0].bar(labels, oa1, width, label='Green in Unpaywall', color='#19905b')
    axs[0].bar(labels, [oa2[k] + oa3[k] for k in range(0, len(oa2))], width,
               label='Gold or Hybrid or Bronze\nin Unpaywall', bottom=oa1, color='#e3bd41')
    
    axs[1].bar(labels, oa1, width, label='hosted only on an open repository', color='#19905B')
    axs[1].bar(labels, oa3, width,
               label='hosted on an open repository\nand by the publisher', bottom=oa1, color='#91AE4F')
    axs[1].bar(labels, oa2, width,
               label='hosted only by the publisher', 
               bottom=[oa1[k] + oa3[k] for k in range(0, len(oa1))], color='#E9D737')
    
    


    # Set number of ticks for x-axis
    ax.set_xticks(range(0, len(oa1)))
    # Set ticks labels for x-axis
    ax.set_xticklabels(labels, rotation='vertical', fontsize=SMALL_SIZE)


    for ax in axs:
        ax.spines.right.set_color('none')
        ax.spines.top.set_color('none')
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_ylim([0, 1])
        ax.set_xlabel('publication year')
        ax.set_ylabel('Open Access rate (%)')
        ax.legend()

    
    axs[0].set_title('Open Access types using\nUnpaywall OA status')
    axs[1].set_title('Open Access types in the\nFrench Open Science Monitor')
    
    fig.savefig('oa_types.png')
```
